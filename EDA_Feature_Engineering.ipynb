```python
df_country=pd.read_excel('Country-code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
#to check dataset
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#pie chart top 3 country zomato used
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

observation:zomato maximun=m or transaction are from india after 
that usa and then united kingdom

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index(name='Rating Count')
```

```python
ratings
```

observation :
1. when rating is between 4.5 to 4.9 -->excellent
2. when the ratings between 4.0 to 4.4 --> very good
3. when rating is between 3.5 to 3.9 -->good
4. when rating between 3.0 to 3.4 --> average
5. when rating is between 2.5 to 2.9 -->average
6. when rating is between 2.0 to 2.4 -->poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,color='green')
```

```python
ratings.head()
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```

observation 
1. Not rated count is very high
2. maximum number of ratings 2.5 to 3.9

```python
#count plot
sns.countplot(x='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```
