```python
#Create 2 D vectors(2D vector)
# plot each vector
#plot v1+v2,v1-v2,v1*4+v1/2
```

```python
vec4=np.array([[1,2],[4,5]])
plt.plot([0,vec4[0][0]],[0,vec4[0][1]],"b--",label="first vector")
plt.plot([0,vec4[1][0]],[0,vec4[1][1]],"r-",label="second vector")
sum=vec4[0][:2]+vec4[1][:2]
plt.plot([0,sum[0]],[0,sum[1]],"g-.",label="sum of v1+bv2")
sub=vec4[0][:2] - vec4[1][:2]
plt.plot([0,sub[0]],[0,sub[1]],"k",label="v1-v2")
res=vec4[0]*4+(vec4[0]/2)
plt.plot([0,res[0]],[0,res[1]],"y",label="v1*4+v1/2")
plt.legend()
plt.show()
```

```python
#matrix Transpose
m1=np.round(10*np.random.randn(3,3))
print(m1)
print(" ")
print(m1.T)
```

```python
m2=np.round(10*np.random.randn(3,3))
print(m2)
print(" ")
trans=m2.T
trans1=trans.T
print(trans)
print(" ")
print(trans1)
#if we transpose the matrix and then again transpose the matrix it will become original matrices
```

```python
m1=np.random.randn(5,4)
m2=np.random.randn(5,4)
print(np.matmul(m1,m2.T))
print(" ")
print(m1@m2.T)
print(" ")
print(np.matmul(m1,m2.T) - m1@m2.T)
```
