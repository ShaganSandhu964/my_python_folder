# Linear regression,Ridge,Lasso

```python
#house pricing dataset
from sklearn.datasets import fetch_california_housing
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=fetch_california_housing()
```

```python
type(df)
```

```python
df
```

```python
dataset=pd.DataFrame(df.data)
```

```python
dataset.columns=df.feature_names
print(dataset.head())
```

```python
#depend features we made is price
dataset['price']=df.target
```

```python
dataset.head()
```

```python
#dividing the dataset into depend and independent features
x=dataset.iloc[:,:-1] #independent
y=dataset.iloc[:,-1] #dependent
```

```python
x.head()
```

```python
y.head()
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
mse=cross_val_score(lin_reg,x,y,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
#here we can give test data and it will predict the value
lin_reg.predict()
```

# Ridge regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}

ridge_regression=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regression.fit(x,y)
```

```python
print(ridge_regression.best_params_)
print(ridge_regression.best_score_)
```

# lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()

params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}

lasso_regression=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regression.fit(x,y)
```

```python
print(lasso_regression.best_params_)
```

```python
print(lasso_regression.best_score_)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
#linear rRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
mse=cross_val_score(lin_reg,x_train,y_train,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge_regression=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regression.fit(x_train,y_train)
```

```python
print(ridge_regression.best_params_)
```

```python
print(ridge_regression.best_score_)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()

params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}

lasso_regression=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regression.fit(x_train,y_train)
```

```python
y_predict=lasso_regression.predict(x_test)
from sklearn.metrics import r2_score
r2_score1=r2_score(y_predict,y_test)
```

```python
print(r2_score1)
```
