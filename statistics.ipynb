# Kurtosis

# probablity Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import bernoulli,gaussian_kde

#probablity
p=0.6
data_bernoulli=bernoulli.rvs(p,size=1000)
plt.hist(data_bernoulli)
plt.show()
```

```python
#this is more to attractive histogram plot
p=0.6
data_bernoulli=bernoulli.rvs(p,size=1000)
plt.hist(data_bernoulli,bins=np.arange(-0.5,2,1),density=True,alpha=0.5,color='pink',edgecolor='black',linewidth=2,align='mid')
plt.show()
```

```python
p=0.6
data_bernoulli=bernoulli.rvs(p,size=1000)
plt.hist(data_bernoulli,bins=np.arange(-0.5,2,1),density=True,alpha=0.5,linewidth=2,color='pink',edgecolor='black',align='mid')
#calculate the kde
kde=gaussian_kde(data_bernoulli)
x=np.linspace(-1,2,300)
y=kde.pdf(x)
#plot the kde 
plt.plot(x,y,color='red')
plt.show()
```

# poisson Distribution

# Continuous Probablity Distribution

# 1. Normal Distribution  2. Uniform Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm,gaussian_kde

mu=1 #mean
sigma=5 # standard deviation
data_norm=norm.rvs(mu,sigma,size=10000)
plt.hist(data_norm,bins=30,density=True,alpha=0.5,color='green')
#calculate the kde
kde=gaussian_kde(data_norm)
x=np.linspace(data_norm.max(),data_norm.min(),301)
y=kde.pdf(x)
plt.plot(x,y,color='red')
# adding labels
plt.xlabel("Normal Distribution")
plt.ylabel("Frequesncy")
plt.show()
```

# Uniform distribution

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import uniform,gaussian_kde

a=0
b=10
data_uni=uniform.rvs(a,b-a,size=10000)
plt.hist(data_uni,bins=30,color='green',density=True,alpha=0.5)

#calculate the kde
kde=gaussian_kde(data_uni)
x=np.linspace(a,b,301)
y=uniform.pdf(x,a,b-a)
plt.plot(x,y,color='red')
#adding plot
plt.xlabel('uniform distribution')
plt.ylabel('Frequency')
plt.show()
```
